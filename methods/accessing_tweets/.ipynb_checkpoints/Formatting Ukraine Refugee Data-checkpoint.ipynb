```python
# Manipulate json files
import json
# Manipulate dataframes
import pandas as pd
# For saving the response data in CSV format
import csv

# Define rootpath
rp = 'C:\\Users\\sgmmahon\\Documents\\GitHub\\iom_project\\'
mp = 'methods\\accessing_tweets\\'
dp = 'data\\tweet_data\\' 

# Define file pathway
pathway = rp + dp + 'tweets/ukraine_refugees_2022.csv'
```

```python
# Load in json file
with open(rp + dp + 'tweets/ukraine-refugees_2022.json') as f:
   data = json.load(f)
```

```python
# Create empty pandas dataframe with correct column titles
migr = pd.DataFrame(columns = pd.concat([pd.DataFrame(data['data'][0]).drop('population_groups', axis=1),
                                         pd.DataFrame(data['data'][0]['population_groups']).drop(columns=['id','name'], axis=1),
                                         pd.DataFrame(data).drop('data', axis=1)], axis=1).columns.values.tolist())

# For loop which extracts data from different levels of json file and appends to migr
for country in data['data']:
    # Extract data at different nested levels into separate dataframes
    cols1 = pd.DataFrame(country).drop('population_groups', axis=1)
    cols2 = pd.DataFrame(country['population_groups']).drop(columns=['id','name'], axis=1)
    cols3 = pd.DataFrame(data).drop('data', axis=1).iloc[0:1]
    # Append data to migr dataframe
    migr = migr.append(pd.concat([cols1, cols2, cols3],axis=1))
    
# Reset dataframe index
migr = migr.rename(columns={'population_groups_concat': 'population_group_name'}).reset_index().drop('index',axis=1)

# Save as csv
migr.to_csv(pathway)

# Show results
migr
```
